# Programa_#01

El programa determina si un número es par o impar.

```
print("Ingrese un número")
value=int(input())
if value%2==0: 
   if value==0:
      print("El número es 0") 
   else:
      print("El número es par")
else:
   print("El número es impar")
```

# Programa_#03

El programa calcula el salario de un trabajador de acuerdo al número de  horas registrado.

```
def Salario(NH,SPH=150):
    SPD=NH*SPH
    SPS=SPD*5
    SPM=SPD*20
    print("El salario por día es:",SPD) 
    print("El salario por semana (5 días) es:",SPS) 
    print("El salario por mes (20 días) es:",SPM) 
Salario(4)
```

# Programa_#05

Una tienda de juguetes tiene dos productos estrella, un payaso y un juego de mesa, el payaso pesa 150 gr y el juego de mesa pesa 170 gr. Hacer una función que reciba el número de payasos y el número de juegos de mesa por pedido, si el peso total del pedido es mayor a 10 kg, debe de mandar error, debe de indicar por cuántos gramos se paso el envío y debe de regresar False al programa principal.

```
P=150
JM=170
print("Ingrese el número de payasos")
NP=int(input()) 
print("Ingrese el número de juegos de mesa") 
NJM=int(input())  
if (NP*P+NJM*JM)>=10000:
   print("Error, exceso de peso") 
   print("El peso excedente es:\n",(NP*P+NJM*JM)-10000,"gr.","\n",((NP*P+NJM*JM)-10000)/1000,"kg")
(NP*P+NJM*JM)<10000
```

# Programa_#06

El programa calcula los impuestos a pagar de acuerdo al salario.

```
print("Ingrese su salario")
S=int(input()) 
if S<10000:
   print("Menos de 10 mil")
   D=S*0.05
   print("Su salario con el descuento es:",S-D)
   print(S<10000) 
else:
   if 10000<=S<=15000:
      print("Entre 10 mil y 15 mil")
      D=S*0.1
      print("Su salario con el descuento es:",S-D)
      print(10000<=S<=15000) 
   else:
      if 15000<S<=20000:
         print("Entre 15 mil y 20 mil")
         D=S*0.15
         print("Su salario con el descuento es:",S-D)
         print(15000<S<=20000)
      else:
         print("Más de 20 mil")
         D=S*0.18
         print("Su salario con el descuento es:",S-D)
         print(S>10000)
```

# Programa_#09

El programa calcula el factorial de un número.

```
print("Ingrese un número entero")
N=int(input())
Cont=N-1
P=N*Cont
while Cont!=1:
   Cont-=1
   P=P*Cont
print(P)
```

 # Programa_#10

El programa calcula el área de un círculo y además el volumen de un cilindro formado por el círculo.

```
Pi=3.14159265
print("Ingrese el radio de la circunferencia") 
r=int(input()) 
A=Pi*(r**2)
print("El área del círculo es:",A)
```

```
print("Ingrese la altura del cilindro")
h=int(input()) 
print("El volumen del cilindro es:",A*h)
```
